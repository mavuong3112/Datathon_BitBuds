# Contact concentration — 5 category BĐS

Phân tích **mức độ tập trung lead (explicit contact)** theo tin, user, kênh và địa lý — **1010 / 1020 / 1030 / 1040 / 1050**.

| Khái niệm | Định nghĩa |
|-----------|------------|
| **Explicit contact** | `view_phone`, `contact_chat`, `contact_zalo`, `contact_sms` |
| **Datathon positive** (tham chiếu) | explicit + `other_interaction` |
| **User metrics** | `is_login = 'login'` |

**Tiền xử lý (README / Cursor.md):** time coverage `dim_listing` vs facts; chỉ login cho user; không gom `other_interaction` vào explicit.

**Chạy cell code** — 1 lần quét `fact_user_events` (~2–3 phút, RAM ~2GB).


```python
%matplotlib inline
from __future__ import annotations

import warnings
from pathlib import Path

import duckdb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import Markdown, display

warnings.filterwarnings("ignore", category=FutureWarning)
sns.set_theme(style="whitegrid", context="notebook")

DATA_ROOT = Path.cwd().resolve()
if not (DATA_ROOT / "dim_listing").exists():
    DATA_ROOT = DATA_ROOT.parent
for name in ("dim_listing", "fact_user_events"):
    if not (DATA_ROOT / name).exists():
        raise FileNotFoundError(f"Thiếu `{name}` trong {DATA_ROOT}")

OUT_DIR = DATA_ROOT / "Thinh_Analyze" / "outputs" / "contact_concentration"
OUT_DIR.mkdir(parents=True, exist_ok=True)

DIM_GLOB = str(DATA_ROOT / "dim_listing" / "*.parquet")
EVENTS_GLOB = str(DATA_ROOT / "fact_user_events" / "*.parquet")
SNAP_GLOB = str(DATA_ROOT / "fact_listing_snapshot" / "*.parquet")

DUCKDB_MEMORY_LIMIT = "2GB"
DUCKDB_THREADS = 2
LOGIN_WHERE = "is_login = 'login'"
EXPLICIT_TYPES = ("view_phone", "contact_chat", "contact_zalo", "contact_sms")
POSITIVE_TYPES = EXPLICIT_TYPES + ("other_interaction",)
EXPLICIT_SQL = ", ".join(repr(x) for x in EXPLICIT_TYPES)
POSITIVE_SQL = ", ".join(repr(x) for x in POSITIVE_TYPES)
CAT_IN = ", ".join(str(c) for c in (1010, 1020, 1030, 1040, 1050))

CAT_META = {
    1010: "1010 — Căn hộ / Chung cư",
    1020: "1020 — Nhà ở",
    1030: "1030 — Văn phòng / Mặt bằng",
    1040: "1040 — Đất",
    1050: "1050 — Phòng trọ",
}
CATEGORIES = tuple(CAT_META)


def gini(values: np.ndarray) -> float:
    x = np.asarray(values, dtype=np.float64)
    x = x[x > 0]
    if x.size == 0:
        return float("nan")
    x = np.sort(x)
    n = x.size
    return float((2 * np.arange(1, n + 1) - n - 1) @ x / (n * x.sum()))


def top_pct_share(values: np.ndarray, top_pct: float) -> float:
    x = np.sort(values)[::-1]
    total = float(x.sum())
    if total == 0:
        return float("nan")
    k = max(1, int(np.ceil(len(x) * top_pct / 100.0)))
    return 100.0 * float(x[:k].sum()) / total


def hhi_shares(shares: np.ndarray) -> float:
    s = np.asarray(shares, dtype=np.float64)
    s = s[s > 0]
    if s.size == 0:
        return float("nan")
    s = s / s.sum()
    return float(10_000 * (s ** 2).sum())


def concentration_row(entity: str, counts: np.ndarray) -> dict:
    n = int(len(counts))
    total = int(counts.sum())
    return {
        "entity": entity,
        "n_with_contact": n,
        "total_contacts": total,
        "mean_contacts": round(total / n, 3) if n else np.nan,
        "median_contacts": float(np.median(counts)) if n else np.nan,
        "gini": round(gini(counts), 4),
        "top_1pct_share": round(top_pct_share(counts, 1), 2),
        "top_5pct_share": round(top_pct_share(counts, 5), 2),
        "top_10pct_share": round(top_pct_share(counts, 10), 2),
        "top_20pct_share": round(top_pct_share(counts, 20), 2),
        "share_from_single_contact_entities_pct": round(100.0 * (counts == 1).sum() / n, 2) if n else np.nan,
    }


con = duckdb.connect(":memory:")
con.execute(f"PRAGMA memory_limit='{DUCKDB_MEMORY_LIMIT}'")
con.execute(f"PRAGMA threads={DUCKDB_THREADS}")

# §0 — time coverage
time_cov = con.execute(
    f"""
    SELECT * FROM (
        SELECT 'dim_listing.posted_date' AS src,
               CAST(MIN(posted_date) AS DATE) AS t_min,
               CAST(MAX(posted_date) AS DATE) AS t_max,
               COUNT(*)::BIGINT AS n_rows
        FROM read_parquet('{DIM_GLOB}')
        UNION ALL
        SELECT 'fact_user_events.date', MIN(date), MAX(date), COUNT(*)::BIGINT
        FROM read_parquet('{EVENTS_GLOB}')
        UNION ALL
        SELECT 'fact_listing_snapshot.date', MIN(date), MAX(date), COUNT(*)::BIGINT
        FROM read_parquet('{SNAP_GLOB}')
    ) ORDER BY 1
    """
).df()
EDA_MIN, EDA_MAX = con.execute(
    f"SELECT MIN(date), MAX(date) FROM read_parquet('{EVENTS_GLOB}')"
).fetchone()

display(Markdown("### §0 — Time coverage (dim vs facts)"))
display(time_cov)
time_cov.to_csv(OUT_DIR / "00_time_coverage.csv", index=False)

qa_contact_flag = con.execute(
    f"""
    SELECT
        SUM(CASE WHEN is_contact = 1 AND event_type IN ({EXPLICIT_SQL}) THEN 1 ELSE 0 END)::BIGINT AS ok_explicit,
        SUM(CASE WHEN is_contact = 1 AND event_type NOT IN ({EXPLICIT_SQL}) THEN 1 ELSE 0 END)::BIGINT AS flag1_non_explicit,
        SUM(CASE WHEN is_contact = 0 AND event_type IN ({EXPLICIT_SQL}) THEN 1 ELSE 0 END)::BIGINT AS explicit_flag0
    FROM read_parquet('{EVENTS_GLOB}')
    WHERE {LOGIN_WHERE}
    """
).df()
display(Markdown("### QA — `is_contact` vs explicit (Nhóm A)"))
display(qa_contact_flag)

# --- Một lần quét events: materialize explicit + positive counts ---
print("Quét fact_user_events (explicit + positive) — một pass…")
con.execute(
    f"""
    CREATE TEMP TABLE explicit_ev AS
    SELECT
        category,
        user_id,
        CAST(item_id AS VARCHAR) AS item_id,
        event_type,
        COALESCE(NULLIF(TRIM(city_name), ''), '(blank)') AS city_name
    FROM read_parquet('{EVENTS_GLOB}')
    WHERE {LOGIN_WHERE}
      AND category IN ({CAT_IN})
      AND event_type IN ({EXPLICIT_SQL})
    """
)

listing_all = con.execute(
    """
    SELECT category, item_id, COUNT(*)::BIGINT AS contacts
    FROM explicit_ev WHERE item_id IS NOT NULL
    GROUP BY 1, 2
    """
).df()

user_all = con.execute(
    "SELECT category, user_id, COUNT(*)::BIGINT AS contacts FROM explicit_ev GROUP BY 1, 2"
).df()

channel_all = con.execute(
    """
    SELECT category, event_type, COUNT(*)::BIGINT AS contacts
    FROM explicit_ev GROUP BY 1, 2 ORDER BY 1, 3 DESC
    """
).df()
channel_all["pct_of_explicit"] = channel_all.groupby("category")["contacts"].transform(
    lambda s: (100.0 * s / s.sum()).round(2)
)

city_all = con.execute(
    """
    SELECT category, city_name, contacts FROM (
        SELECT category, city_name, COUNT(*)::BIGINT AS contacts,
               ROW_NUMBER() OVER (PARTITION BY category ORDER BY COUNT(*) DESC) AS rn
        FROM explicit_ev GROUP BY 1, 2
    ) WHERE rn <= 10
  ORDER BY category, contacts DESC
    """
).df()

pos_by_cat = con.execute(
    f"""
    SELECT category, COUNT(*)::BIGINT AS datathon_positive_events
    FROM read_parquet('{EVENTS_GLOB}')
    WHERE {LOGIN_WHERE} AND category IN ({CAT_IN}) AND event_type IN ({POSITIVE_SQL})
    GROUP BY 1
    """
).df().set_index("category")["datathon_positive_events"]

supply = con.execute(
    f"""
    SELECT
        category,
        COUNT(*)::BIGINT AS listings_in_dim,
        SUM(CASE WHEN posted_date BETWEEN DATE '{EDA_MIN}' AND DATE '{EDA_MAX}' THEN 1 ELSE 0 END)::BIGINT AS listings_posted_in_eda_window
    FROM read_parquet('{DIM_GLOB}')
    WHERE category IN ({CAT_IN})
    GROUP BY 1
    """
).df().set_index("category")

seller_all = con.execute(
    f"""
    WITH ec AS (
        SELECT category, item_id, COUNT(*)::BIGINT AS contacts
        FROM explicit_ev WHERE item_id IS NOT NULL GROUP BY 1, 2
    )
    SELECT ec.category, d.seller_id, SUM(ec.contacts)::BIGINT AS contacts
    FROM ec
    INNER JOIN read_parquet('{DIM_GLOB}') d
        ON ec.item_id = CAST(d.item_id AS VARCHAR) AND ec.category = d.category
    WHERE d.seller_id IS NOT NULL
    GROUP BY 1, 2
    """
).df()

summary_rows: list[dict] = []

for cat in CATEGORIES:
    label = CAT_META[cat]
    print(f"\n{'=' * 60}\n{label}\n{'=' * 60}")

    listing_counts = listing_all.loc[listing_all["category"] == cat, "contacts"].to_numpy()
    user_counts = user_all.loc[user_all["category"] == cat, "contacts"].to_numpy()
    seller_counts = seller_all.loc[seller_all["category"] == cat, "contacts"].to_numpy()

    conc_listing = concentration_row("listing", listing_counts)
    conc_user = concentration_row("user_login", user_counts)
    conc_seller = concentration_row("seller", seller_counts)

    supply_dim = int(supply.loc[cat, "listings_in_dim"]) if cat in supply.index else 0
    supply_eda = int(supply.loc[cat, "listings_posted_in_eda_window"]) if cat in supply.index else 0
    listings_with_contact = conc_listing["n_with_contact"]
    pct_catalog = round(100.0 * listings_with_contact / supply_dim, 2) if supply_dim else np.nan

    channel = channel_all[channel_all["category"] == cat].copy()
    ch_hhi = round(hhi_shares(channel["contacts"].to_numpy()), 1) if len(channel) else np.nan

    city = city_all[city_all["category"] == cat].copy()
    city_total = float(city["contacts"].sum()) if len(city) else 0.0
    city_top5 = round(100.0 * city.head(5)["contacts"].sum() / city_total, 2) if city_total else np.nan
    city["pct_of_top10_shown"] = (100.0 * city["contacts"] / city_total).round(2) if city_total else 0

    pos_events = int(pos_by_cat.get(cat, 0))

    row = {
        "category": cat,
        "category_label": label,
        "listings_in_dim": supply_dim,
        "listings_posted_in_eda_window": supply_eda,
        "listings_with_explicit_contact": listings_with_contact,
        "pct_listings_with_contact_vs_dim": pct_catalog,
        "total_explicit_contacts": conc_listing["total_contacts"],
        "contacts_per_listing_in_dim": round(conc_listing["total_contacts"] / supply_dim, 4) if supply_dim else np.nan,
        "listing_gini": conc_listing["gini"],
        "listing_top_1pct_share": conc_listing["top_1pct_share"],
        "listing_top_10pct_share": conc_listing["top_10pct_share"],
        "user_gini": conc_user["gini"],
        "user_top_10pct_share": conc_user["top_10pct_share"],
        "seller_gini": conc_seller["gini"],
        "seller_top_10pct_share": conc_seller["top_10pct_share"],
        "dominant_channel": channel.iloc[0]["event_type"] if len(channel) else None,
        "channel_hhi_x1e4": ch_hhi,
        "city_top5_share_pct": city_top5,
        "datathon_positive_events": pos_events,
        "explicit_share_of_positive_pct": round(100.0 * conc_listing["total_contacts"] / pos_events, 2) if pos_events else np.nan,
    }
    summary_rows.append(row)

    detail = pd.DataFrame([conc_listing, conc_user, conc_seller])
    detail.insert(0, "category", cat)
    detail.to_csv(OUT_DIR / f"{cat}_concentration_by_entity.csv", index=False)

    display(Markdown(f"**{label}**"))
    display(detail)
    display(channel[["event_type", "contacts", "pct_of_explicit"]])
    display(city[["city_name", "contacts", "pct_of_top10_shown"]])

summary = pd.DataFrame(summary_rows)
summary.to_csv(OUT_DIR / "01_summary_all_categories.csv", index=False)
channel_all.to_csv(OUT_DIR / "02_channel_mix_by_category.csv", index=False)
city_all.to_csv(OUT_DIR / "03_city_top10_by_category.csv", index=False)

display(Markdown("## Tổng hợp 5 category"))
display(summary)

fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))
x = np.arange(len(summary))
w = 0.35
axes[0].bar(x - w / 2, summary["listing_gini"], width=w, label="Listing Gini", color="#2171b5")
axes[0].bar(x + w / 2, summary["user_gini"], width=w, label="User Gini", color="#238b45")
axes[0].set_xticks(x)
axes[0].set_xticklabels(summary["category"].astype(str))
axes[0].set_ylabel("Gini")
axes[0].set_title("Tập trung explicit contact")
axes[0].legend(fontsize=8)

axes[1].bar(x - w / 2, summary["listing_top_10pct_share"], width=w, label="Top 10% listings", color="#cb181d")
axes[1].bar(x + w / 2, summary["user_top_10pct_share"], width=w, label="Top 10% users", color="#6a51a3")
axes[1].set_xticks(x)
axes[1].set_xticklabels(summary["category"].astype(str))
axes[1].set_ylabel("% contacts")
axes[1].set_title("Pareto top 10%")
axes[1].legend(fontsize=8)
plt.tight_layout()
fig.savefig(OUT_DIR / "fig_01_gini_and_top10_share.png", dpi=120, bbox_inches="tight")
plt.show()

piv = channel_all.pivot(index="category", columns="event_type", values="pct_of_explicit").fillna(0).reindex(CATEGORIES)
piv.plot(kind="bar", stacked=True, figsize=(9, 4.5), colormap="Set2", edgecolor="white")
plt.ylabel("% explicit contacts")
plt.title("Mix kênh liên hệ (login)")
plt.xticks(rotation=0)
plt.legend(title="event_type", bbox_to_anchor=(1.02, 1), fontsize=8)
plt.tight_layout()
plt.savefig(OUT_DIR / "fig_02_channel_mix.png", dpi=120, bbox_inches="tight")
plt.show()

print("DATA_ROOT =", DATA_ROOT)
print("Export →", OUT_DIR)
```

## §1 — Trùng bài sau cắt time coverage (`dim_listing`)

**Cắt tập:** `posted_date ∈ [min(events.date), max(events.date)]` — đồng bộ với cửa sổ fact (Cursor.md: time coverage bất đồng bộ).

**Không** nhóm theo `item_id` (mỗi `item_id` vẫn unique — Nhóm A). Kiểm tra **trùng nội dung** qua fingerprint:

| Mức | Khóa (sau chuẩn hóa chữ) | Ý nghĩa |
|-----|--------------------------|---------|
| **L1** | `title` + `category` | Tiêu đề giống hệt (nới nhất) |
| **L2** | L1 + `ad_type`, `city_name`, `district_name`, `price_bucket` | Trùng “cùng tin” theo primary fields (role matrix) |
| **L3** | L2 + `seller_id`, `ward_name`, `area_sqm` (làm tròn) | Gần như cùng một bài đăng |
| **L4** | `title` + `category` + `seller_id` | Cùng seller repost tiêu đề |

Chuẩn hóa: `lower`, `trim`, gộp khoảng trắng (Nhóm D — whitespace).

```python
DUP_OUT = DATA_ROOT / "Thinh_Analyze" / "outputs" / "dim_duplicate_after_timecut"
DUP_OUT.mkdir(parents=True, exist_ok=True)

# EDA_MIN / EDA_MAX đã có từ cell trên; chạy độc lập thì uncomment 2 dòng sau:
# EDA_MIN, EDA_MAX = con.execute(f"SELECT MIN(date), MAX(date) FROM read_parquet('{EVENTS_GLOB}')").fetchone()

display(Markdown(f"**Cửa sổ cắt:** `posted_date` từ **{EDA_MIN}** → **{EDA_MAX}**"))

con.execute(
    f"""
    CREATE OR REPLACE TEMP TABLE dim_scoped AS
    SELECT
        CAST(item_id AS VARCHAR) AS item_id,
        seller_id,
        category,
        ad_type,
        seller_type,
        posted_date,
        expected_expired_date,
        title,
        city_name,
        district_name,
        ward_name,
        price_bucket,
        area_sqm,
        bedrooms,
        bathrooms,
        regexp_replace(lower(trim(coalesce(title, ''))), '\\s+', ' ', 'g') AS title_norm,
        regexp_replace(lower(trim(coalesce(city_name, ''))), '\\s+', ' ', 'g') AS city_norm,
        regexp_replace(lower(trim(coalesce(district_name, ''))), '\\s+', ' ', 'g') AS district_norm,
        regexp_replace(lower(trim(coalesce(ward_name, ''))), '\\s+', ' ', 'g') AS ward_norm,
        regexp_replace(lower(trim(coalesce(price_bucket, ''))), '\\s+', ' ', 'g') AS price_norm
    FROM read_parquet('{DIM_GLOB}')
    WHERE posted_date BETWEEN DATE '{EDA_MIN}' AND DATE '{EDA_MAX}'
      AND title IS NOT NULL AND trim(title) <> ''
    """
)

scope_stats = con.execute(
    """
    SELECT
        COUNT(*)::BIGINT AS n_listings_scoped,
        COUNT(DISTINCT item_id)::BIGINT AS n_distinct_item_id,
        COUNT(*) - COUNT(DISTINCT item_id)::BIGINT AS duplicate_item_id_rows,
        COUNT(DISTINCT title_norm)::BIGINT AS n_distinct_title_norm,
        MIN(posted_date) AS min_posted,
        MAX(posted_date) AS max_posted
    FROM dim_scoped
    """
).df()
display(Markdown("### Baseline sau cắt (kiểm tra `item_id` chỉ để đối chiếu Nhóm A)"))
display(scope_stats)

LEVELS = {
    "L1_title_category": "title_norm, category",
    "L2_title_geo_price": "title_norm, category, ad_type, city_norm, district_norm, price_norm",
    "L3_title_full_fingerprint": (
        "title_norm, category, ad_type, seller_id, city_norm, district_norm, ward_norm, "
        "price_norm, round(coalesce(area_sqm, -1), 1)"
    ),
    "L4_title_seller": "title_norm, category, seller_id",
}

level_rows = []
for level_name, group_cols in LEVELS.items():
    row = con.execute(
        f"""
        WITH grp AS (
            SELECT {group_cols},
                   COUNT(*)::BIGINT AS n_items,
                   COUNT(DISTINCT item_id)::BIGINT AS n_distinct_item_ids,
                   COUNT(DISTINCT seller_id)::BIGINT AS n_sellers,
                   COUNT(DISTINCT posted_date)::BIGINT AS n_posted_dates,
                   MIN(title) AS title_example
            FROM dim_scoped
            GROUP BY ALL
            HAVING COUNT(*) > 1
        )
        SELECT
            '{level_name}' AS dup_level,
            (SELECT COUNT(*) FROM dim_scoped)::BIGINT AS n_listings_scoped,
            COUNT(*)::BIGINT AS duplicate_groups,
            COALESCE(SUM(n_items), 0)::BIGINT AS rows_in_dup_groups,
            COALESCE(SUM(n_items - 1), 0)::BIGINT AS excess_rows_beyond_first,
            ROUND(100.0 * COALESCE(SUM(n_items), 0) / (SELECT COUNT(*) FROM dim_scoped), 2) AS pct_rows_in_dup_groups,
            COALESCE(SUM(CASE WHEN n_sellers = 1 THEN 1 ELSE 0 END), 0)::BIGINT AS groups_same_seller,
            COALESCE(SUM(CASE WHEN n_sellers > 1 THEN 1 ELSE 0 END), 0)::BIGINT AS groups_multi_seller,
            COALESCE(MAX(n_items), 0)::BIGINT AS max_group_size
        FROM grp
        """
    ).df().iloc[0].to_dict()
    level_rows.append(row)

dup_summary = pd.DataFrame(level_rows)
dup_summary.to_csv(DUP_OUT / "01_duplicate_levels_summary.csv", index=False)
display(Markdown("### Tổng hợp trùng nội dung (không dùng `item_id`)"))
display(dup_summary)

# Theo category — dùng L2 (khuyến nghị đọc)
by_cat = con.execute(
    """
    WITH grp AS (
        SELECT category, title_norm, ad_type, city_norm, district_norm, price_norm,
               COUNT(*)::BIGINT AS n_items,
               COUNT(DISTINCT seller_id)::BIGINT AS n_sellers
        FROM dim_scoped
        GROUP BY ALL
        HAVING COUNT(*) > 1
    )
    SELECT
        category,
        COUNT(*)::BIGINT AS duplicate_groups,
        SUM(n_items)::BIGINT AS rows_in_dup_groups,
        SUM(n_items - 1)::BIGINT AS excess_rows,
        ROUND(100.0 * SUM(n_items) / (SELECT COUNT(*) FROM dim_scoped d WHERE d.category = grp.category), 2) AS pct_category_rows_in_dup,
        SUM(CASE WHEN n_sellers = 1 THEN 1 ELSE 0 END)::BIGINT AS groups_same_seller,
        SUM(CASE WHEN n_sellers > 1 THEN 1 ELSE 0 END)::BIGINT AS groups_multi_seller
    FROM grp
    GROUP BY category
    ORDER BY category
    """
).df()
by_cat["category_label"] = by_cat["category"].map(CAT_META)
by_cat.to_csv(DUP_OUT / "02_L2_duplicate_by_category.csv", index=False)
display(Markdown("### L2 — trùng theo category (`title` + geo + giá + `ad_type`)"))
display(by_cat)

# Top nhóm trùng (L2) + phân loại same vs multi seller
top_groups = con.execute(
    """
    WITH grp AS (
        SELECT
            category, ad_type, city_norm, district_norm, price_norm,
            title_norm,
            MIN(title) AS title_example,
            COUNT(*)::BIGINT AS n_items,
            COUNT(DISTINCT seller_id)::BIGINT AS n_sellers,
            COUNT(DISTINCT posted_date)::BIGINT AS n_posted_dates,
            MIN(posted_date) AS first_posted,
            MAX(posted_date) AS last_posted,
            CASE WHEN COUNT(DISTINCT seller_id) = 1 THEN 'same_seller_repost'
                 ELSE 'multi_seller_clone' END AS dup_kind
        FROM dim_scoped
        GROUP BY ALL
        HAVING COUNT(*) > 1
    )
    SELECT * FROM grp ORDER BY n_items DESC LIMIT 25
    """
).df()
top_groups.to_csv(DUP_OUT / "03_L2_top25_duplicate_groups.csv", index=False)
display(Markdown("### Top 25 nhóm trùng L2 (ví dụ thực tế)"))
display(top_groups)

# Chi tiết 1 nhóm mẫu: multi-seller, nhiều bản nhất
example = top_groups.iloc[0]
detail = con.execute(
    f"""
    SELECT item_id, seller_id, seller_type, posted_date, expected_expired_date,
           ad_type, city_name, district_name, ward_name, price_bucket, area_sqm, bedrooms, title
    FROM dim_scoped
    WHERE category = {int(example['category'])}
      AND title_norm = ?
      AND ad_type = ?
      AND city_norm = ?
      AND district_norm = ?
      AND price_norm = ?
    ORDER BY posted_date
    LIMIT 15
    """,
    [
        example["title_norm"],
        example["ad_type"],
        example["city_norm"],
        example["district_norm"],
        example["price_norm"],
    ],
).df()
display(Markdown(
    f"### Drill-down nhóm lớn nhất: `{example['title_example'][:80]}…` "
    f"({int(example['n_items'])} tin, {int(example['n_sellers'])} seller)"
))
display(detail)

# Chart: % rows in dup groups by level
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(dup_summary["dup_level"], dup_summary["pct_rows_in_dup_groups"], color="#756bb1")
ax.set_xlabel("% listings (scoped) nằm trong nhóm trùng")
ax.set_title("Mức độ trùng nội dung sau cắt time coverage")
for i, v in enumerate(dup_summary["pct_rows_in_dup_groups"]):
    ax.text(v + 0.1, i, f"{v:.2f}%", va="center", fontsize=9)
plt.tight_layout()
fig.savefig(DUP_OUT / "fig_duplicate_levels.png", dpi=120, bbox_inches="tight")
plt.show()

fig, ax = plt.subplots(figsize=(8, 4))
x = np.arange(len(by_cat))
w = 0.35
ax.bar(x - w / 2, by_cat["groups_same_seller"], width=w, label="Cùng seller", color="#3182bd")
ax.bar(x + w / 2, by_cat["groups_multi_seller"], width=w, label="Nhiều seller", color="#e6550d")
ax.set_xticks(x)
ax.set_xticklabels(by_cat["category"].astype(str))
ax.set_ylabel("Số nhóm trùng (L2)")
ax.set_title("L2 — repost vs clone đa seller theo category")
ax.legend()
plt.tight_layout()
fig.savefig(DUP_OUT / "fig_L2_same_vs_multi_seller.png", dpi=120, bbox_inches="tight")
plt.show()

print("Export →", DUP_OUT)
```
